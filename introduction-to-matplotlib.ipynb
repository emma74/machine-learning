## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10))  #(width, length) of fig

# 3. plot data
ax.plot(x, y)
plt.show()

# 4. customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save and show (you show the whole data)
fig.savefig("images/sample-data.png")

```

## Making figures with Numpy array


```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2)
plt.show()
```

```python
# Use the same data to make a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
plt.show()
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
plt.show()
```

```python
# Bar plot grap
butter_prices = {"peanut butter":8,
                 "Almond butter": 10,
                 "Cashew Butter": 12}

fig, ax = plt.subplots()
ax.bar(butter_prices.keys(), butter_prices.values())
ax.set(title="Dan's Butter Store",
       ylabel="Prices ($)")
plt.show()
```

```python
#Horizontal bar
fig, ax = plt.subplots()
ax.barh(list(butter_prices.keys()), list(butter_prices.values()))
plt.show()
```

```python
# Make data for hostogram and plot it 
x = np.random.randn(1000) #generate 1000 random numbers between 0 and 1
fx, ax = plt.subplots()
ax.hist(x, density=True)
plt.show()
```

## Two options for subplots

```python
# subplot option 1
fg, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))

# plot to each axis
ax1.plot(x, x/2)
ax2.scatter(np.random.rand(10), np.random.rand(10))
ax3.bar(butter_prices.keys(), butter_prices.values())
ax3.set(ylabel="Prices ($)")

ax4.hist(np.random.randn(1000))
                                                   
plt.show()
```

```python
# subplots option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))

# plot to each different axis
ax[0, 0].plot(x, x/2)  #top-left subplot
ax[0, 1].scatter(np.random.rand(10), np.random.rand(10)) #top-right subplot
ax[1, 0].bar(butter_prices.keys(), butter_prices.values()) #bottom-left subplot
ax[1, 0].set(xlabel="Prices ($)")
ax[1, 1].hist(np.random.randint(1000), density=True)
plt.show()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
len(over_50)
```

```python
type(heart_disease["target"][0])
```

```python
# pyplot method
heart_disease.plot(kind='scatter',
                   x='age',
                   y='chol',
                   c='target')

plt.show()
```

```python
## OO method with pyplot
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',x='age',y='chol',c='target',ax=ax)
#ax.set_xlim(45, 100)  #set x-axis to a wider range
plt.show()
```

```python
#OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# plot data
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])

# customize the plot
ax.set(title="Heart and Cholestrol Levels",
       ylabel="Cholestrol",
       xlabel="Age")

# Add a legend
ax.legend(*scatter.legend_elements(), title="target")

#Add broken horizontal line to mark the mean
ax.axhline(over_50["chol"].mean(),
           linestyle='--')
       
plt.show()
```

```python
heart_disease
```

```python
# create subplots: age, cholestrol and thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])

# Customize ax0
ax0.set(title="Heart Disease and Cholestrol Levels",
        ylabel="Cholestrol")

#Add a legend
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a mean line
ax0.axhline(y=over_50["chol"].mean(), linestyle="--")

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")

#Add a legend
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a mean line
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--")

# Add title to figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")

plt.show()
```
