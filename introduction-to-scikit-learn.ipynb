## 0. An end-to-end Scikit-Learn workflow

## 1. Getting data ready

```python
# Get the data ready
import pandas as pd
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
# create Y lables
y = heart_disease["target"]
y
```

X.shape, y.shape

```python
# Create X (features matrix)
X = heart_disease.drop("target", axis=1)
X
```

```python
# 2. Choose the right model and hyperparameters
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()

# We will keep the default parameters
clf.get_params()
```

```python
# Fit the model into training data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
clf.fit(X_train, y_train)
```

```python
X_test
```

```python
y_test
```

```python
y_train
```

```python
X_test
```

```python
# Make a prediction. Hint data to be used should have the same shape as data used to train (X_train)
y_preds = clf.predict(X_test)
y_preds
```

```python
# Evaluate the model on train 
clf.score(X_train, y_train)
```

```python
# Evaluate the model on test data
clf.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test, y_preds)
```

```python
# 5. Improve a model
# Try different amounts of n_estimators
np.random.seed(42)
for i in range(100, 200, 10):
    print(f"Trying model with {i} estimators...")
    clf = RandomForestClassifier(n_estimators=i).fit(X_train, y_train)
    print(f"Model accuracy on the test set: {clf.score(X_test, y_test) * 100:.2f}%")
    print(" ")
```

```python
# 6. Save a model and load it
import pickle

pickle.dump(clf, open("random_forest_model_1.pkl", "wb"))
```

```python
loaded_model = pickle.load(open("random_forest_model_1.pkl", "rb"))
loaded_model.score(X_test, y_test)
```
